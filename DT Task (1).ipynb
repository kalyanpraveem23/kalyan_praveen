```python
from sklearn.preprocessing  import LabelEncoder
```

```python
x = df_1.drop('Temp', axis=1)
```

```python
y = df_1['Temp']
y
```

```python
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_absolute_error, r2_score
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.25,random_state=45)
```

```python
x_train.shape
```

```python
model = DecisionTreeRegressor()
```

```python
model.fit(x_train,y_train)
```

```python
y_train_pred=model.predict(x_train)
```

```python
y_train_pred
```

```python
mean_absolute_error(y_train,y_train_pred)
```

```python
r2_score(y_train,y_train_pred)
```

```python
from sklearn.ensemble import RandomForestRegressor
```

```python
model_2 = RandomForestRegressor()
```

```python
model_2.fit(x_train,y_train)
```

```python
y_pred_model=model_2.predict(x_train)
mean_absolute_error(y_train,y_pred_model)
```

```python
y_pred_model=model_2.predict(x_train)
r2_score(y_train,y_pred_model)
```

```python
y_pred_model=model_2.predict(x_test)
mean_absolute_error(y_test,y_pred_model)
```
