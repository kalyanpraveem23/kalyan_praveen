## 1 Univariate

### 1Univariate Continous data---Histogram

```python
plt.figure(figsize =(10,5))
plt.hist(x="",data=mt_cars_df)
plt.xlabel('')
plt.ylabel('Freq')
plt.title("MPG DISTRIBUTION",size=15)
```

```python
plt.figure(figsize =(10,5))
plt.hist(x="disp",data=mt_cars_df)
plt.xlabel('disp')
plt.ylabel('Freq')
plt.title("disp Distribution",size=15)
plt.show()
```

```python
plt.boxplot(x="mpg",data=mt_cars_df)
plt.show()
```

### Univariate Discrete data--- Count-Plot

```python
plt.bar(x="cyl",data=mt_cars_df)
```

```python
sns.countplot(x="cyl",data=mt_cars_df)
plt.show()
```

```python
plt.pie(x=mt_cars_df["cyl"].value_counts(),autopct='%1.0f%%',labels=[8,4,6],explode=[0.02,0,0.03])
```
