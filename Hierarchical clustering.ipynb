```python
import pandas as pd
import numpy as np
from matplotlib import pyplot as pt
import seaborn as sns
from sklearn.cluster import AgglomerativeClustering
import scipy.cluster.hierarchy as sch
```

```python
university_df=pd.read_excel("Universities.xlsx")
```

```python
university_df
```

```python
def norm_fun(i):
    x=(i-i.min())/(i.max()-i.min())
    return(x)
```

```python
nor_df=norm_fun(university_df.iloc[:,1:])
```

```python
nor_df
```

```python
dendogram=sch.dendrogram(sch.linkage(nor_df))
```

```python
hc=AgglomerativeClustering(n_clusters=5,metric="euclidean")
```

```python
y = hc.fit_predict(nor_df)
```

```python
y
```

```python
university_df["Cluster"]=y
```
