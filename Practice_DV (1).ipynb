### Univariate Continous data -Histogram

```python
plt.hist(ai_dev_df['ai_usage_hours’])
```

### Univariate Discrete data ---CountPlot

```python
sns.countplot(x='distractions',data=ai_dev_df)
plt.show()
```

```python
plt.figure(figsize=(4,4))  
sns.countplot(x='task_success', data=ai_dev_df)
plt.title("Task Success Count")           
plt.xlabel("Task Success (0 = Fail, 1 = Success)")  
plt.ylabel("Number of Tasks")             
plt.show()
```

```python
plt.figure(figsize=(8,5))  
sns.countplot(x='bugs_reported',hue='bugs_reported', data=ai_dev_df, palette='pastel')  
plt.title("Bug Reporting Frequency Among Developers", fontsize=14)  
plt.xlabel("Number of Bugs Reported")
plt.ylabel("Number of Developers")
plt.tight_layout()
plt.show()
```

### Piechart for Discrete data 

```python
task_counts = ai_dev_df["task_success"].value_counts().sort_index()
```

```python
plt.pie(x=ai_dev_df["task_success"].value_counts(),autopct='%1.0f%%',labels=task_counts.index.astype(str),explode=[0.02,0.02],startangle=90)
plt.title('Task Success vs Failure Distribution')
plt.show()
```

```python
bug_counts = ai_dev_df["bugs_reported"].value_counts().sort_index()
```

```python
plt.figure(figsize=(10,4))
sns.scatterplot(x='hours_coding', y='sleep_hours', hue='task_success', markers=True, data=ai_dev_df)
plt.title("Coding Hours vs Sleep by Task Success")
plt.gca().invert_xaxis()
plt.show()

```

```python
plt.figure(figsize=(8,5))
sns.scatterplot(x='sleep_hours', y='coffee_intake_mg', hue='task_success', data=ai_dev_df)
plt.title("Coffee Intake vs Sleep Hours by Task Success")
plt.xlabel("Sleep Hours")
plt.ylabel("Coffee Intake (mg)")
plt.tight_layout()
plt.show()
```

```python
plt.title("AI Usage vs Coding Hours by Task Success")
sns.scatterplot(x='hours_coding',y='ai_usage_hours',hue='task_success',data=ai_dev_df)
plt.show()
```

```python
sns.barplot(x='distractions',y='commits',hue='task_success',data=ai_dev_df)
plt.show()
```

```python
sns.barplot(x='task_success',y='ai_usage_hours',hue='task_success',data=ai_dev_df)
plt.show()
```

```python
plt.figure(figsize=(8,6))  
sns.barplot(
    x='distractions',
    y='commits',
    hue='task_success',
    data=ai_dev_df,
    palette='Set2',
    ci=None,
    order=sorted(ai_dev_df['distractions'].unique())  # Optional: Sort x-axis categories
)
plt.title("Impact of Distractions on Commits by Task Success", fontsize=14)
plt.xlabel("Distractions Level")
plt.ylabel("Average Number of Commits")
plt.legend(title="Task Success\n(0 = Fail, 1 = Success)")
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,4))
sns.scatterplot(x='hours_coding', y='sleep_hours', hue='task_success', markers=True, data=ai_dev_df)
plt.title("Coding Hours vs Sleep by Task Success")
plt.gca().invert_xaxis()
plt.show()


```

```python
sns.barplot(x='distractions', y='hours_coding', hue='task_success', data=ai_dev_df)
plt.title("Coding Hours by Distraction Type and Task Success")
plt.ylabel("Average Coding Hours")
plt.xlabel("Type of Distraction")
plt.show()
```
