```python
MOD = 10**9 + 7

# Memoization dictionary for divisor counts
divisor_cache = {}

def count_divisors(n):
    if n in divisor_cache:
        return divisor_cache[n]
    count = 0
    i = 1
    while i * i <= n:
        if n % i == 0:
            count += 1
            if i != n // i:
                count += 1
        i += 1
    divisor_cache[n] = count
    return count

def get_count(N, X, A):
    dp = [0] * (N + 1)
    dp[0] = 1

    for i in range(1, N + 1):
        product = 1
        for j in range(i, 0, -1):
            product *= A[j - 1]
            if product > 1e12:
                break
            if count_divisors(product) >= X:
                dp[i] = (dp[i] + dp[j - 1]) % MOD
    return dp[N]

# Test cases
N = 3
X = 5
A = [2, 6, 8]
print(get_count(N, X, A))

N2 = 2
X2 = 2
A2 = [2, 2]
print(get_count(N2, X2, A2))
```

```python
MOD = 10**9 + 7

def get_count(n, x, a):
    memo = {}
    def dfs(idx):
        if idx == n:
            return 1
        if idx in memo:
            return memo[idx]

        res = 0
        prod = 1
        for j in range(idx, n):
            prod *= a[j]
            if prod > 1e12:  # Optional cutoff
                break
            if count_factors(prod) >= x:
                res = (res + dfs(j + 1)) % MOD
        memo[idx] = res
        return res

    return dfs(0)

def count_factors(num):
    cnt = 0
    i = 1
    while i * i <= num:
        if num % i == 0:
            cnt += 1
            if i != num // i:
                cnt += 1
        i += 1
    return cnt

# Example usage:
n1, x1, a1 = 3, 5, [2, 6, 8]
print(get_count(n1, x1, a1))  # Output: 1

n2, x2, a2 = 2, 2, [2, 2]
print(get_count(n2, x2, a2))  # Output: 2
```
