```python
import pandas as pd
from mlxtend.frequent_patterns import apriori,association_rules
```

```python
retail_data=pd.read_excel("Online Retail.xlsx")
```

```python
France_data=retail_data[retail_data['Country']=="France"]
```

```python
France_data
```

```python
France_data.isna().sum()
```

```python
France_data.info()
```

```python
df_france=pd.pivot_table(data=France_data,values="Quantity",index='InvoiceNo', columns="Description").fillna(0)
df_france
```

```python
encoder_france=df_france.applymap(lambda x:1 if x>0 else 0)
encoder_france
```

```python
frequent_item=apriori(encoder_france,min_support=0.05,use_colnames=True)
frequent_item
```

```python
best_association = association_rules(frequent_item)
```
